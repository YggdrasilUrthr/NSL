```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
liquid_T_ne = np.loadtxt("NonEq/Liquid/output_temp_1.dat")

plt.errorbar(liquid_T_ne[:,0], liquid_T_ne[:,2], yerr=liquid_T_ne[:,3])
plt.grid(True)
plt.show()
```

```python
ex_liquid_T = 1.1
ob_liquid_T = liquid_T_ne[-1,2]

offset = ex_liquid_T - ob_liquid_T
new_temp = ex_liquid_T + offset

print("Offset: " + offset.astype(str))
print("New Liquid Temp: " + new_temp.astype(str))
```

```python
directory_name = "Eq/Liquid/"
directory = os.fsencode(directory_name)
    
for file in os.listdir(directory):

    filename = os.fsdecode(file)
    data = np.loadtxt(directory_name + filename)

    plt.errorbar(data[20:,0], data[20:,2], yerr=data[20:,3])
    plt.grid(True)
    plt.title(filename)
    plt.show()    
```
